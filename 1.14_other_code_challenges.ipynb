```python
from itertools import permutations


def words(letters):
    for n in range(1, len(letters) + 1):
        yield from map(''.join, permutations(letters, n))

for word in words("BAnana"):
    print(word)
```

```python
# moves the zeros in a list to the end of the list
def move_zeros(lst):
    return [x for x in lst if x != 0] + [0] * lst.count(0)


def move_zeros2(array):
    return sorted(array, key=lambda x: not x)


move_zeros([1, 0, 1, 2, 0, 1, 3])
```

```python
# A Narcissistic Number (or Armstrong Number) is a positive number which is the sum of its own digits,
# each raised to the power of the number of digits in a given base.
# 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153


def narcissistic(value):
    ex = map(int, str(value))
    p = len(str(value))
    n = [x**p for x in ex]
    print(n)
    return sum(n) == value


def narcissistic2(value):
    return value == sum(int(x)**len(str(value)) for x in str(value))


narcissistic(371)
```

```python
"""Convert a string to a new string where each character in the new string is "(" if that character appears
only once in the original string, or ")" if that character appears more than once in the original string. 
Ignore capitalization when determining if a character is a duplicate."""


def duplicate_encode(word):
    low = word.lower()
    par_lst = ["(" if low.count(w) == 1 else ")" for w in low]
    return "".join(par_lst)


def duplicate_encode2(word):
    low = word.lower()
    par_lst = ["(" if low.count(w) == 1 else ")" for w in low]
    return "".join(par_lst)


duplicate_encode('recede')
```
